```python
# 지도학습모델 : 단순회귀분석(선형회귀모델, 선형회귀예측모델)
# - 종속변수가 있어요
# - 직선의 방정식을 사용 ; y = ax + b
# - y : 종속변수(정답)
# - x : 독립변수
# - a : 기울기
# - b : y절편

# - 학습 : fit
```

```python
# 사용할 변수들 : mpg,sylinders, horsepower, weight
nbf = df[["mpg","cylinders", "horspower", "weight"]]
```

```python
nbf
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.scatter(nbf.mpg, nbf.weight, color = "plum", alpha=0.5)
```

```python
nbf.plot(kind = "scatter",
        x = "weight",
        y = "mpg",
        c = "hotpink",
        s = 10,
        figsize = (10,5))
plt.show()
plt.close()
```

```python
# 해석
# 중량이 커질수록 연비는 낮아지는 음의 상관관계
# 중량은 연비에 영향을 미친다
# 회귀선(추세선) 그리기
import seaborn as sns
nbf.plot(kind = "scatter",
        x = "weight",
        y = "mpg",
        c = "hotpink",
        s = 10,
        figsize = (10,5))
sns.regplot(x='weight', y='mpg', ci=None, data=nbf)
plt.show()
```

```python
# 회귀선(추세선) 그리기
fig = plt.figure(figsize=(10,5))

# 첫번째 그래프 공간 만들기(1행2열의 1열)
ax1 = fig.add_subplot(1,2,1)

# 두번째 그래프 공간 만들기(1행 2열의 2열)
ax2 = fig.add_subplot(1,2,2)

# 첫번째 그래프
sns.regplot(x = "weight",
            y = "mpg" ,
            data = nbf,
            ax = ax1,
            color = "purple",
            fit_reg = True)

# 두번째 그래프
sns.regplot(x = "weight",
            y = "mpg" ,
            data = nbf,
            ax = ax2,
            color = "olive",
            fit_reg = False)
plt.show()
plt.close()
```

```python
sns.jointplot(x = nbf["weight"],
               y = nbf["mpg"])


sns.jointplot(x = nbf["weight"],
               y = nbf["mpg"] ,
             kind = "reg",
             color = 'green',
             marker = 'x')
```
