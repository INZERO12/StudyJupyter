##  단일변수 데이터
###  데이터프레임 변수 유형
### 
- 범주형 변수 : 연속형 데이터와 범주형 데이터의 가장 큰 차이는 가감승제가 가능하지만 의미가 없다는 것
- 질적 자료 : 숫자들의 크기가 차이가 계산되지 않음(명목변수, 서열변수)
### 
- 연속형 변수 : 사칙연산이 가능
- 양적자료 : 숫자들의 크기 차이 계산(등간변수, 비율변수)

### 연속형분석 
#### - 수집된 수치 데이터의 정리, 표현, 요약 등을 통해 데이터의 전반적인 특성을 이해하는 분석 진입 단계, '기술통계'라고도 한다.
#### - 자료의 특성을  찾는 분석 접근 방법

### tips 데이터셋 읽어들이기

```python
# 라이브러리 불러오기
import pandas as pd

import matplotlib as mpl
import matplotlib.pylab as plt
import seaborn as sns

import numpy as np
```

```python
# tips 데이터셋 읽어 들이기
tips = sns.load_dataset("tips")
tips
```

### 빈도분석

```python
# 성별 관련 데이터 뽑기
tips[["sex"]]
```

```python
# 성별 빈도분석
tips["sex"].value_counts()
```

```python
# smoker 빈도분석
tips["smoker"].value_counts()
```

```python
# 요일 빈도분석
tips["day"].value_counts()
```

```python
# time 빈도분석
tips["time"].value_counts()
```

### 교차분석

```python
# 월별/성별 교차분석
pd.crosstab(tips["sex"],tips["day"])
```

```python
# 성별/ 월별 교차분석 총괄표
pd.crosstab(tips["sex"],
            tips["day"], 
            margins=True)
```

```python
# 빈도 비율 확인
pd.crosstab(tips["sex"],
            tips["day"]).apply(
                lambda r: r/len(tips),
                axis = 1
            )
```
